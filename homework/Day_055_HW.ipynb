#### 用 iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn import datasets

np.random.seed(5)

%matplotlib inline
```

```python
iris = datasets.load_iris()
df_iris = pd.DataFrame(iris.data, columns = iris.feature_names)
X = iris.data
y = iris.target
print(df_iris.shape, np.bincount(y))
df_iris.head()
```

```python
# 秀出資料欄位的類型與數量
dtype_df = df_iris.dtypes.reset_index() # 重置 index 值
dtype_df.columns = ["Count", "Column Type"]
dtype_df = dtype_df.groupby("Column Type").aggregate('count').reset_index()
dtype_df
```

* 實驗測試不同的初始值
* 呈現結果

#### 資料建模並視覺化結果

```python
def plot_3Dcluster(X, y, columns, labels, titles, fignum):
    fig = plt.figure(figsize = (8, 6))
    ax = Axes3D(fig, rect = [0, 0, .95, 1])
    ax.scatter(X[:, 1], X[:, 2], X[:, 3], c = columns.astype(np.float), edgecolor='k')
    ax.set_xlabel(labels[0])
    ax.set_ylabel(labels[1])
    ax.set_zlabel(labels[2])
    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    ax.view_init(elev = 15, azim = 225) # elev 垂直旋轉; azim 水平旋轉    
```

```python
estimators = [('k_means_2', KMeans(n_clusters = 2)),
              ('k_means_3', KMeans(n_clusters = 3)),
              ('k_means_4', KMeans(n_clusters = 4)),
              ('k_means_bad_init', KMeans(n_clusters = 5, n_init = 1, init = 'random'))]
```

```python
fignum = 1
titles = ['2 clusters', '3 clusters', '4 clusters', '5 clusters, bad initialization', 'Ground Truth']
for cluster, kms in estimators:
    kms.fit(X)
    columns = kms.labels_
    labels = iris.feature_names[1:]    
    plot_3Dcluster(X, y, columns, labels, titles, fignum) 
    fignum = fignum + 1
```

```python
fig = plt.figure(figsize = (8, 6))
ax = Axes3D(fig, rect = [0, 0, .95, 1])
for name, label in [('cls0', 0),
                    ('cls1', 1),
                    ('cls2', 2)]:
    ax.text3D(X[y == label, 1].mean(),
              X[y == label, 2].mean(),
              X[y == label, 3].mean(), name,
              horizontalalignment = 'center',
              bbox = dict(alpha=.2, edgecolor='w', facecolor='w'))
ax.scatter(X[:, 1], X[:, 2], X[:, 3], c = y, edgecolor='k')
ax.set_xlabel(labels[0])
ax.set_ylabel(labels[1])
ax.set_zlabel(labels[2])
ax.set_title(titles[fignum - 1])
ax.dist = 12
ax.view_init(elev = 15, azim = 225) # elev 垂直旋轉; azim 水平旋轉    
```
